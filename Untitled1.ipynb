```python
import numpy as np
import matplotlib.pyplot as plt
na = 1 #空氣折射率
ng = 1.5 #稜鏡折射率
A = 30*np.pi/180 #稜鏡的頂角

theta_1 = np.linspace(0,90,90)*np.pi/180 #入射角1
theta_1p = np.arcsin(na*np.sin(theta_1)/ng) #折射角1
theta_2 = A-theta_1p #入射角2
theta_2p = np.arcsin(ng*np.sin(theta_2)/na) #折射角2
D = theta_1+theta_2p-A #偏向角
D_min = np.pi 
i = 0
for i in range(90):
    if D[i] < D_min:
        D_min = D[i]
    else:
        D_min = D_min
        #print(i," ",D_min)
    i = i+1
print(D.min())
print(min(D))

plt.plot(theta_1*180/np.pi,D*180/np.pi,"o-")
plt.grid()
```
